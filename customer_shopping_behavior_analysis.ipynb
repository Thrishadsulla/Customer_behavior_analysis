```python
import pandas as pd
df=pd.read_csv("customer_shopping_behavior.csv")
```

```python
df["Review Rating"]=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','')
df=df.rename(columns={'purchaseamount(usd)':'purchaseamount'})
```

```python
df.columns
```

```python
#create a column age_group
labels=['Young Adult','Adult','Middle-aged','Senior']
df['agegroup']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','agegroup']].head(10)
```

```python
#create column purchase frequency days
frequencymapping={
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90
}
df['purchasefrequencydays']=df['frequencyofpurchases'].map(frequencymapping)
```

```python
df[['purchasefrequencydays','frequencyofpurchases']].head(10)
```
